# Aplicando instruções em SQL

**Descrição do Case**: Você foi contratado por uma empresa fictícia para analisar seus dados e fornecer insights que possam ajudar a empresa a entender melhor seus colaboradores.

### Quantos registros possui o banco lotação?

```python
query = """
select
'Nro de Registros' as Nome,
count(*) as qtde
from fato_lotacao
"""
ps.sqldf(query, locals())
```

O banco possui 13 registros.

### Departamento com maior folha salarial?

```python
display(dim_departamento.head(2))
display(fato_lotacao.head(2))
```

```python
query = """
select
dp.nome as departamento,
sum(lt.salario) as maior_salario_dp
from fato_lotacao as lt
inner join dim_departamento as dp on dp.id=lt.id_departamento
group by
dp.nome

"""
ps.sqldf(query, locals())
```

O departamento com maior folha salarial é **Tecnologia da Informação**

### Cargo melhor remunerado (em média)?

```python
display(dim_cargo.head(2))
display(fato_lotacao.head(2))
```

```python
query = """
select
cg.nome as cargo,
avg(lt.salario) as media_salario
from fato_lotacao as lt
inner join dim_cargo as cg on cg.id=lt.id_cargo
group by
cg.nome
order by  media_salario desc
"""
ps.sqldf(query, locals())
```

O cargo de analista de dados é, em média, melhor remunerado.
